```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

image.png

```python
df = pd.read_csv("YOUTUBE CHANNELS DATASET (1).csv")
```

```python
df
```

```python
df.info()
```

```python
# drop the duplicate values
df.drop_duplicates(inplace = True)
```

```python
df["Country"].unique()
```

```python
#droping the null values
df.dropna(inplace = True)
```

```python
country_dict = {
    'US': 'United States',
    'IN': 'India',
    'JP': 'Japan',
    'KR': 'South Korea',
    'CA': 'Canada',
    'PT': 'Portugal',
    'AE': 'United Arab Emirates',
    'BR': 'Brazil',
    'AR': 'Argentina',
    'PK': 'Pakistan',
    'MX': 'Mexico',
    'IT': 'Italy',
    'GB': 'United Kingdom',
    'RU': 'Russia',
    'ES': 'Spain',
    'ID': 'Indonesia',
    'CL': 'Chile',
    'PH': 'Philippines',
    'TR': 'Turkey',
    'PR': 'Puerto Rico',
    'SV': 'El Salvador',
    'CO': 'Colombia',
    'NO': 'Norway'
}
df["Country"] = df["Country"].map(country_dict)
```

```python
df
```

```python
df.reset_index(inplace = True)
```

```python
df
```

```python
df.drop(["index"],axis = 1,inplace = True)
```

```python
df
```

```python
df.info()
```

```python
df["Views"] = df["Views"].str.replace(",","")
```

```python
df
```

```python
df["Views"] = df["Views"].astype(float)
```

```python
df.info()
```

```python
df["Uploads"] = df["Uploads"].str.replace(",","")
```

```python
df["Uploads"] = df["Uploads"].astype(float)
```

```python
df["Subscribers"] = df["Subscribers"].str.strip("M")
```

```python
df["Subscribers"] = df["Subscribers"].astype(float)
```

```python
df
```

```python
#@title top youtubers subscriber count
plt.figure(figsize=(15,5))
a = sns.barplot(x = "Username",y ="Subscribers",data = df.head(10), color = "red", edgecolor = "black")
for i in a.containers:
  a.bar_label(i)

plt.xticks(rotation = 90)
plt.title("Top 10 Youtubers",fontsize = 15)
plt.xlabel("Channel name",fontsize = 13)
plt.ylabel("Total subscriber",fontsize = 13)
plt.show()
```

```python
#@title Total uploads by channels
plt.figure(figsize=(15,5))
a = sns.barplot(x = "Username",y = "Uploads",data = df.sort_values(by = "Uploads",ascending = False).head(10),color = "red",edgecolor = "black")
for i in a.containers:
  a.bar_label(i)
plt.xticks(rotation = 90)
plt.title("top 10 most uploading channels",fontsize = 12)
plt.xlabel("Channel name",fontsize = 12)
plt.ylabel("Uploads",fontsize = 12)

plt.show()
```

```python
#@title top 10 channels  their total views
plt.figure(figsize=(15,5))
sns.barplot(x = "Username",y = "Views", data = df.sort_values(by = "Views",ascending = False).head(10),color = "red",edgecolor = "black")
plt.xticks(rotation = 90)

plt.title("Top 10 YouTubers by Views",fontsize = 13)
plt.xlabel("Channel_name",fontsize = 13)
plt.ylabel("Total Views",fontsize = 13)

plt.show()
```

```python
plt.figure(figsize = (15,6))

d = df["Country"].value_counts().sort_values(ascending = False)
filtered_df = df[df["Country"].isin(d.index)]
a = sns.countplot(x='Country', data=filtered_df, order=d.index ,color = "red",edgecolor = "black")
for i in a.containers:
  a.bar_label(i)

plt.xticks(rotation = 90)
plt.title("Country Representation of Top Creators")
plt.xlabel("Countries")
plt.ylabel("count")

plt.tight_layout()

#sns.despine()
plt.show()

#India, Usa are having top most number of to creators
```

```python
plt.figure(figsize = (15,7))
a = sns.barplot(x = "Country",y = "Views", data = df.sort_values(by = "Views",ascending = False), color = "red",edgecolor = "black",estimator = np.sum,ci=None)
for i in a.containers:
  a.bar_label(i, rotation=90, padding=5)

plt.title("Country with the most views",fontsize = 18)
plt.xlabel("Country",fontsize = 17)
plt.ylabel("Views",fontsize = 17)

plt.xticks(rotation = 90)
plt.tight_layout()
sns.despine()
plt.show()
```

```python
#@title Country with most uploads
plt.figure(figsize=(15,7))
a = sns.barplot(x = "Country",y = "Uploads",data = df.sort_values(by = "Uploads",ascending = False),color = "red",edgecolor = "black",estimator=np.sum,ci=None)
for i in a.containers:
  a.bar_label(i)
plt.xticks(rotation = 90)
plt.title("top 10 most uploading channels",fontsize = 12)
plt.xlabel("Channel name",fontsize = 12)
plt.ylabel("Uploads",fontsize = 12)

plt.tight_layout()
sns.despine()

plt.show()
```
